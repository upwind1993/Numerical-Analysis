```python
import numpy as np
w = 10
y0 = 5
y = 15
x = 50

def f(TA):
    return y-(TA/w*np.cosh(w/TA*x)+y0-TA/w)

from scipy.optimize import brentq

TA = brentq(f,1200,1300)

def g(x):
    return TA/w*np.cosh(w/TA*x)+y0-TA/w

x = np.linspace(-150,100,100)
import pylab
pylab.plot(x,g(x),c='k')
pylab.grid()
pylab.xlabel('x')
pylab.ylabel('y')
```
